```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Define model hyperparameters
input_shape = (224, 224, 3)
num_classes = 17
dropout_rate = 0.2
width = 1
depth = 1.0
batch_size = 32
num_epochs = 50

# Load pre-trained model from .h5 file
pretrained_model = keras.models.load_model('/Users/hemalatha/Downloads/efficientnet_with_cbam_fuused.h5')

# Freeze the layers in the pretrained model
for layer in pretrained_model.layers:
    layer.trainable = False

# Create new model using pretrained layers and new output layer
new_model = keras.models.Sequential()
new_model.add(pretrained_model)
new_model.add(keras.layers.Dense(num_classes, activation='softmax'))  # add new output layer

# Compile the new model
new_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Load new dataset
train_dir = '/Users/hemalatha/Downloads/FYPDataset3/train'
test_dir = '/Users/hemalatha/Downloads/FYPDataset3/test'

train_datagen = ImageDataGenerator(rescale=1./255, rotation_range=20, width_shift_range=0.2, height_shift_range=0.2, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_datagen = ImageDataGenerator(rescale=1./255)

resolution = input_shape[0]

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(resolution, resolution),
    batch_size=batch_size,
    class_mode='categorical',
    subset='training')

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(resolution, resolution),
    batch_size=batch_size,
    class_mode='categorical')

# Train the new output layer
new_model.fit(train_generator, epochs=num_epochs, batch_size=batch_size)

# Evaluate the model
test_loss, test_accuracy = new_model.evaluate(test_generator)
print('Test accuracy:', test_accuracy)

# Save the model
new_model.save('disease_efficientnet_with_cbam_fuused.h5')
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator


# Load the pre-trained model from .h5 file
model = keras.models.load_model('NEXT_FINAL_CROPDISEASE_efficientnet_with_cbam_fuused.h5')

# Load the test dataset
test_dir = '/Users/hemalatha/Downloads/FYPDataset3/test'
test_datagen = ImageDataGenerator(rescale=1./255)
test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(224, 224),
    batch_size=1,
    class_mode='categorical',
    shuffle=False
)

# Evaluate the model
test_loss, test_accuracy = model.evaluate(test_generator)
print('Test accuracy:', test_accuracy)

# Predict labels for the test data
predictions = model.predict(test_generator)
predicted_labels = np.argmax(predictions, axis=1)
true_labels = test_generator.classes

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)
print('Confusion Matrix:')
print(cm)
```
